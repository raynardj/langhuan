## Load data

```python
news20 = fetch_20newsgroups()
df = pd.DataFrame(dict(text=news20['data']))
df.head()
```

```python
df["newidx"] = list(f"i_{i}" for i in range(len(df)))
df = df.set_index("newidx")
```

```python
!rm -rf ~/.cache/langhuan/test_case
```

```python
app = NERTask.from_df(df,
                      text_col="text",
                      task_name="test_case",
                      options=["institution", "company", "name"],
                      load_history=True
                     )
```

```python
app.run("localhost", port=5566)
```
